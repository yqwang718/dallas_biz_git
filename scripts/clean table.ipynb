```python
import pandas as pd
import os
import re
print("Current working directory:", os.getcwd())
print("Files in current directory:", os.listdir('.'))

```

```python
# Load the data
df = pd.read_csv('offenders_biz_summary.csv')
# df = pd.read_csv('victims_biz_summary.csv')
df.head()
```

```python
def clean_format(text):
    """Transform the format more explicitly"""
    if pd.isna(text) or not isinstance(text, str):
        return text
    
    # Split by ± to get mean and std part
    if '±' in text:
        parts = text.split('±')
        mean = parts[0].strip()
        
        # Extract std (before the parentheses)
        std_part = parts[1].strip()
        std = std_part.split('(')[0].strip()
        
        return f"{mean} ({std})"
    else:
        return text

# Apply to all data columns (skip Category and Field)
data_columns = ['burglary_breaking_entering', 'motor_vehicle_theft', 
                'larceny_theft_offenses', 'assault_offenses', 'robbery', 
                'drug_narcotic_violations', 'all_combined']

for col in data_columns:
    df[col] = df[col].apply(clean_format)

print(df.head())
```
